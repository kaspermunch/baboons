```python
meta_data_samples = pd.read_csv("../data/Papio_metadata_with_clustering.txt", sep =" ")
```

```python
meta_data_samples_Sci = meta_data_samples.copy()
for i, row in meta_data_samples_Sci.iterrows():
    if row.PGDP_ID[0] != "P":
        meta_data_samples_Sci.at[i, "PGDP_ID"] = "Sci_"+str(row.PGDP_ID)
meta_data_samples = meta_data_samples_Sci
```

```python
meta_data_samples.C_origin.unique()
```

```python
reference_meta = meta_data_samples.loc[meta_data_samples.C_origin.isin([
       'Kindae, Zambia', 'Gelada, Captive', 'Hamadryas, Ethiopia',
       'Papio, Senegal', 'Ursinus, Zambia'
])]
```

```python
query_meta = meta_data_samples.loc[meta_data_samples.C_origin.isin([
       'Cynocephalus, Central Tanzania', 'Anubis, Kenya',
       'Anubis, Tanzania', 'Cynocephalus, Western Tanzania',
       'Anubis, Ethiopia'
])]
```

```python
females_reference = reference_meta.loc[reference_meta.Sex == "F"]
```

```python
females_query = query_meta.loc[(query_meta.Sex == "F") & (query_meta.Origin != "Aberdare, Kenya")]
#Removing Aberdare as it only is present in low numbers (2 samples) and in the old dataset
```

```python
females_reference
```

```python
females_query
```

```python
output_dir = "/home/eriks/baboondiversity/data/PG_panu3_phased_chromosomes_4_7_2021/rfmix_input/"
females_reference.PGDP_ID.to_csv(output_dir+"females_reference_samples.txt", index=False, header=False)
females_query.PGDP_ID.to_csv(output_dir+"females_query_samples.txt", index=False, header=False)
reference_meta.PGDP_ID.to_csv(output_dir+"reference_samples.txt", index=False, header=False)
query_meta.PGDP_ID.to_csv(output_dir+"query_samples.txt", index=False, header=False)
```
